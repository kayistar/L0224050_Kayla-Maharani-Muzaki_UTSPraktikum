**b.Buat 1 model regresi linear untuk memprediksi nilai variabel dependen Sales dengan variabel independen TV**

```python
X = df [['TV']].values
y = df ['Sales'].values
```

```python
print (y)
print (X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_test, y_test, X_train, y_train = train_test_split(X, y, test_size = 0.2, random_state = 0 )

print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
print(y_train.shape)
```

Train test split tidak saya gunakan karena errorm, maka saya akan menggunakan data asli

```python
y_pred_false = 300 + 40 * X_test

print(y_pred_false)
```

```python
from sklearn.linear_model import LinearRegression 
linreg = LinearRegression()
linreg.fit(X, y)
y_pred_true = linreg.predict(X)
print(y_pred_true)
```

**Tentukan slope dan intercept dari model tersebut.Jelaskan maksud dari slope dan interceptnya**

```python
print("Intercept;", linreg.intercept_)
print("Slope;", linreg.coef_)
```

**Hitung nilai R-square, MSE, dan RMSE dari model yang telah dibuat. Berikan analisis dari kinerja model tersebut.**

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error(y, y_pred_true)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred_true)

print(f"mse: {mse}")
print(f"rmse: {rmse}")
print(f"r-square : {r2}")
```

**ANALISIS**
<br> setelah mencari y prediksi (asli), selanjutnya adalah mencari mse, rmse dan r-squarenya. Dapat dilihat diatas bahwa msenya adalah 10.512652915656757, brarti nilai error dari data ini sebesar 10.512652915656757.

```python
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def model_dinamik (r, p, P, K) :
    dPdt = r * P * (1 -p/K)
    return dPdt

# deklarasi 
r = 0.80
P0 = 200
K = 5000
t = np.linspace (0, 100)

sol = odeint (model_dinamik, P0, t, args = (r, K))

print(sol)
```
